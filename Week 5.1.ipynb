```
from pyspark.sql.types import*
mySchema =  StructType ([StructField("index", IntegerType()) ,
                         StructField("webpage", StringType()) ,
                         StructField("associated_files", StringType())])
                         
myRDD= sc.textFile("/FileStore/webpage/*")\
                         .map(lambda line: line.split("\t"))\
                         .map(lambda values:[int(values[0]), values[1], values[2]])

webpages= spark.createDataFrame(myRDD, mySchema)   
                         
webpages.show(truncate=False)  
```

```
webpages = spark.read.options(delimiter = "\t").csv("/FileStore/webpage/*")
```

```
webpages.printSchema()
```

```
webpages.show(4)
```

```
namedw=webpages.withColumnRenamed("_c0","num").withColumnRenamed("_c1", "webpage").withColumnRenamed("_c2","associated_files")
webpages= namedw.withColumn("num", namedw["num"].cast("int"))
webpages.printSchema()
```

```
webpages.show(4)
```

```
assocfilesDF=webpages.select(webpages.webpage, webpages.associated_files)
from pyspark.sql import DataFrame
isinstance(assocfilesDF, DataFrame)
```

```
assocfilesDF.show()
```

```
afilesRDD= assocfilesDF.rdd.map(lambda row: (row.webpage, row.associated_files))
```

```
afilesRDD.take(4)
```

```
afilesRDD2= afilesRDD.flatMapValues(lambda filestring: filestring.split(","))
afilesRDD2.take(4)
```

```
afileDF= spark.createDataFrame(afilesRDD2)
afileDF.show()
```

```
afileDF.printSchema()
```

```
finalDF= afileDF.withColumnRenamed("_1", "web_page_num").withColumnRenamed("_2", "associated_file")
```

```
finalDF.printSchema()
```

```
try:
    dbutils.fs.ls("/FileStore/webpage_files/")
    dbutils.fs.rm("/FileStore/webpage_files/", True)
except:
    pass
finalDF.write.save("/FileStore/webpage_files/")
```
